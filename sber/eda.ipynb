```python
import pandas as pd
df = pd.read_csv('happinessu2024.csv',encoding='latin')
df.info()
```

```python
 print(f"Размер датасета: {df.shape}")
 print("\nТипы данных:\n", df.dtypes)
 print("\nПропуски:\n", df.isnull().sum())
```

```python
# Средний уровень счастья по годам
mean_happiness = df.groupby('year')['Life Ladder'].mean()
print(mean_happiness)

# Процент стран с высоким уровнем свободы
high_freedom = df[df['Freedom to make life choices'] > 0.7]['Country name'].nunique() / df['Country name'].nunique() * 100
print(f"Доля стран с высоким уровнем свободы: {high_freedom:.2f}%")
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Корреляционная матрица
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица')

plt.show()
```

```python
import pandas as pd
df = pd.read_csv('happinessu2024.csv',encoding='latin')
df.info()
```

```python
import plotly.express as px

fig = px.histogram(df, x='Life Ladder', title='Распределение уровня счастья')
fig.show()
```

```python
mean_happiness = df.groupby('year')['Life Ladder'].mean().reset_index()
fig = px.line(mean_happiness, x='year', y='Life Ladder', title='Динамика уровня счастья')
fig.show()
```

```python
latest_year = df['year'].max()
top_countries = df[df['year'] == latest_year][['Country name', 'Life Ladder']].nlargest(10, 'Life Ladder')
fig = px.bar(top_countries, x='Country name', y='Life Ladder', title=f'Топ-10 стран по уровню счастья ({latest_year})')
fig.show()
```

```python
# df.info()
fig = px.scatter(df, x='Log GDP per capita', y='Life Ladder', color='year', hover_data=['Country name'],
                 title='Связь ВВП на душу населения и уровня счастья')
fig.show()
```

```python
# df.info()
# Метрики для карточек
mean_happiness = df['Life Ladder'].mean()
median_life_expectancy = df['Healthy life expectancy at birth'].median()
high_happiness_countries = df[df['Life Ladder'] > 6]['Country name'].nunique()

print(f"Средний уровень счастья: {mean_happiness:.2f}")
print(f"Медиана продолжительности жизни: {median_life_expectancy:.2f}")
print(f"Количество стран с высоким уровнем счастья: {high_happiness_countries}")
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('happinessu2024.csv', encoding='latin')

# Основная информация
print(f"Размер датасета: {df.shape}")
print("\nПропуски:\n", df.isnull().sum())
print("\nОписательные статистики:\n", df.describe())

# Обработка пропусков
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# Метрики
mean_happiness = df.groupby('year')['Life Ladder'].mean().reset_index()
high_freedom = df[df['Freedom to make life choices'] > 0.7]['Country name'].nunique() / df['Country name'].nunique() * 100

# Визуализации
# 1. Распределение уровня счастья
fig1 = px.histogram(df, x='Life Ladder', title='Распределение уровня счастья')
fig1.show()

# 2. Динамика счастья по годам
fig2 = px.line(mean_happiness, x='year', y='Life Ladder', title='Динамика уровня счастья')
fig2.show()

# 3. Топ-10 стран
latest_year = df['year'].max()
top_countries = df[df['year'] == latest_year][['Country name', 'Life Ladder']].nlargest(10, 'Life Ladder')
fig3 = px.bar(top_countries, x='Country name', y='Life Ladder', title=f'Топ-10 стран ({latest_year})')
fig3.show()

# 4. Корреляция
fig4 = px.scatter(df, x='Log GDP per capita', y='Life Ladder', color='year', hover_data=['Country name'],
                  title='Связь ВВП и уровня счастья')
fig4.show()

# Корреляционная матрица
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица')
plt.show()
```
